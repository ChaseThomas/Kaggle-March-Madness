# generate 2017 Model input data

```python
import pandas as pd
pd.options.display.max_rows=200
pd.options.display.max_columns=50
```

```python
df_season=pd.read_csv('2017_Final_DetailedResults.csv')
df_season
```

```python
df_seed=pd.read_csv('TourneySeeds_Thru2017_Better.csv')
df_seed=df_seed[df_seed['Season']==2017]
df_seed
```

```python
df_season['Wfg3_percentage']=df_season['Wfgm3']/df_season['Wfga3']
df_season['Lfg3_percentage']=df_season['Lfgm3']/df_season['Lfga3']
df_season['Wfg_percentage']=df_season['Wfgm']/df_season['Wfga']
df_season['Lfg_percentage']=df_season['Lfgm']/df_season['Lfga']
df_season['Wor_percentage']=df_season['Wor']/df_season['Wfga']
df_season['Lor_percentage']=df_season['Lor']/df_season['Lfga']
df_season['Wdr_percentage']=df_season['Wdr']/df_season['Lfga']
df_season['Ldr_percentage']=df_season['Ldr']/df_season['Wfga']
df_season
```

```python
a=df_season[['Season', 'Wteam', 'Wfg3_percentage', 'Wfg_percentage', 'Wor_percentage', 'Wdr_percentage']]
a.columns=['Season', 'team', 'fg3_percentage', 'fg_percentage', 'or_percentage', 'dr_percentage']
b=df_season[['Season', 'Lteam', 'Lfg3_percentage', 'Lfg_percentage', 'Lor_percentage', 'Ldr_percentage']]
b.columns=['Season', 'team', 'fg3_percentage', 'fg_percentage', 'or_percentage', 'dr_percentage']
c=pd.concat([a,b])
#df_seaon_final['Team']=df_seaon_final.index
#df_seaon_final
```

```python
df_season_final=c.groupby(['Season','team']).mean()
df_season_final['temp']=df_season_final.index
df_season_final[['Season', 'team']]=df_season_final.temp.apply(pd.Series)
df_season_final
```

```python
#result=
df_seed_final=pd.merge(df_seed, df_season_final, how='inner', left_on=['Season', 'Team'], right_on=['Season', 'team'])
df_seed_final['Seed']=df_seed_final['Seed'].str.extract('(\d+)').astype(int)
df_seed_final=df_seed_final.iloc[:,:7]
df_seed_final
#df['B'].str.extract('(\d+)').astype(int)
```

```python
from itertools import product
df_allCompact=pd.DataFrame(list(product(df_seed_final['Team'], df_seed_final['Team'])), columns=['Team1', 'Team2'])
df_allCompact=df_allCompact[df_allCompact.Team1!=df_allCompact.Team2]
df_allCompact=df_allCompact[df_allCompact.Team1<df_allCompact.Team2]
df_allCompact
```

```python
df_compactResult_final=pd.merge(df_allCompact, df_seed_final, how='inner', left_on='Team1', right_on='Team')
df_compactResult_final=pd.merge(df_compactResult_final, df_seed_final, how='inner', left_on='Team2', right_on='Team')
df_compactResult_final
```

```python
df_compactResult_final['W_L_Seed']=df_compactResult_final['Seed_x']-df_compactResult_final['Seed_y']
df_compactResult_final['W_L_fg3_percentage']=df_compactResult_final['fg3_percentage_x']-df_compactResult_final['fg3_percentage_y']
df_compactResult_final['W_L_fg_percentage']=df_compactResult_final['fg_percentage_x']-df_compactResult_final['fg_percentage_y']
df_compactResult_final['W_L_or_percentage']=df_compactResult_final['or_percentage_x']-df_compactResult_final['or_percentage_y']
df_compactResult_final['W_L_dr_percentage']=df_compactResult_final['dr_percentage_x']-df_compactResult_final['dr_percentage_y']
df_compactResult_final
```

```python
df_result=df_compactResult_final[['Season_x','Team1', 'Team2', 'W_L_Seed','W_L_fg3_percentage','W_L_fg_percentage', 'W_L_or_percentage', 'W_L_dr_percentage']]
df_result=df_result.rename(columns={'Season_x': 'Season'})
df_result
#df_temp
```

```python
df_result.to_csv('Model_input_2017.csv', index=False)
```
