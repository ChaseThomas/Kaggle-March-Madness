```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

## Logistic Regression

```python
df_input=pd.read_csv('Model_input.csv', index_col=False)
df_input
```

```python
year=2003

df_test=df_input[df_input['Season']==year]
df_train=df_input[df_input['Season']!=year]


X_test=df_test.iloc[:,3:8]
y_test=df_test['label']

X_train=df_train.iloc[:,3:8]
y_train=df_train['label']
```

```python

#df_input=pd.read_csv('Model_input_labelZeroOne.csv', index_col=False)
#df_input


#X=df_input[['W-L_Seed', 'W-L_fg3_percentage']]
#y=df_input['label']

logistic = LogisticRegression()
logistic.fit(X_train, y_train)

print 'coef.: ', logistic.coef_

#misclassify rate
print 'misclassifiation rate(train): ', sum(logistic.predict(X_train)*y_train<0)/float(len(y_train))
print 'misclassifiation rate(test): ', sum(logistic.predict(X_test)*y_test<0)/float(len(y_test))

log_preds=logistic.predict_log_proba(X_test)
y_true=y_test
y_true[y_true==-1]=0

#Evaluation
print 'Evalutation: ', -(y_true*log_preds[:,0]+(1-y_true)*log_preds[:,1]).mean()
```

## Decision Tree

```python
#df_input=pd.read_csv('Model_input_labelZeroOne.csv', index_col=False)
#df_input

#X=df_input[['W-L_Seed', 'W-L_fg3_percentage']]
#y=df_input['label']

clf = DecisionTreeClassifier(max_depth=3)
clf.fit(X_train, y_train)

print '.: ', clf.n_outputs_

#misclassify rate
print 'misclassifiation rate(train): ', sum(clf.predict(X_train)*y_train<0)/float(len(y_train))
print 'misclassifiation rate(t#est): ', sum(clf.predict(X_test)*y_test<0)/float(len(y_test))

#feature_importances_
print 'feature_importances_: ', clf.feature_importances_

log_preds=clf.predict_log_proba(X_test)
y_true=y_test
y_true[y_true==-1]=0
#Evaluation
print 'Evalutation: ', -(y_true*log_preds[:,1]+(1-y_true)*log_preds[:,0]).mean()

tree.export_graphviz(clf,out_file='tree.dot') 
#log_preds
#clf.classes_
```

# Output the training set for the ensemble method

```python
df_result=df_input.copy()
for year in range(2014,2017):
    
    df_test=df_input[df_input['Season']==year]
    df_train=df_input[df_input['Season']!=year]


    X_test=df_test.iloc[:,3:8]
    y_test=df_test['label']

    X_train=df_train.iloc[:,3:8]
    y_train=df_train['label']
    
    depth=3
    if year==2014:
        depth=2

    clf = DecisionTreeClassifier(max_depth=depth)
    clf.fit(X_train, y_train)

    print 'Season: ', year

    #misclassify rate
    print 'misclassifiation rate(train): ', sum(clf.predict(X_train)*y_train<0)/float(len(y_train))
    print 'misclassifiation rate(test): ', sum(clf.predict(X_test)*y_test<0)/float(len(y_test))

    #feature_importances_
    print 'feature_importances_: ', clf.feature_importances_

    log_preds=clf.predict_log_proba(X_test)
    y_true=y_test
    y_true[y_true==-1]=0
    #Evaluation
    print 'Evalutation: ', -(y_true*log_preds[:,1]+(1-y_true)*log_preds[:,0]).mean()

    preds=clf.predict_proba(X_test)
    preds=pd.DataFrame(preds)[1]
    preds.index=X_test.index
    df_result.loc[df_result['Season']==year,'pred_prob']=pd.DataFrame(preds)[1]
df_result
```
